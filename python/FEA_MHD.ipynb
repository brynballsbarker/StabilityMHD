```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

$$
\frac{1}{\Delta t} \int \rho v + \int (\rho u)_x v = \frac{1}{\Delta t} \int \rho_0 v
\tag{1}
$$

$$
\frac{1}{\Delta t} \int \rho u v + \int (\rho u^2-h^2)_x v + \int (p + \frac{1}{2}h^2)_xv - \mu \int u_{xx} v = \frac{1}{\Delta t} \int \rho_0 u_0 v
\tag{2}
$$

$$
\frac{1}{\Delta t}\int  \left(\rho e + \frac{1}{2}(\rho u^2+h^2)\right)v + \int 
\left((\rho e + \frac{1}{2}\rho u^2 + p)u \right)_xv - \int \left( (\eta+2\mu)uu_x \right)_x v = \kappa \int T_{xx} v + \frac{1}{\Delta t} \int \left(\rho_0 e_0 + \frac{1}{2}(\rho_0u_0^2+h_0^2)\right)v
\tag{3}
$$

$$
\frac{1}{\Delta t}\int h v - \nu \int h_{xx} v = \frac{1}{\Delta t} \int h_0 v
\tag{4}
$$

Now apply integration by parts to get

$$
\frac{1}{\Delta t} \int \rho v - \int \rho u v_x = \frac{1}{\Delta t} \int \rho_0 v
\tag{1}
$$

$$
\frac{1}{\Delta t} \int \rho u v - \int (\rho u^2-h^2) v_x - \int (p + \frac{1}{2}h^2)v_x + \mu \int u_{x} v_x = \frac{1}{\Delta t} \int \rho_0 u_0 v
\tag{2}
$$

$$
\frac{1}{\Delta t}\int  \left(\rho e + \frac{1}{2}(\rho u^2+h^2)\right)v - \int 
\left((\rho e + \frac{1}{2}\rho u^2 + p)u \right)v_x + \int \left( (\eta+2\mu)uu_x \right) v_x = - \kappa \int T_{x} v_x + \frac{1}{\Delta t} \int \left(\rho_0 e_0 + \frac{1}{2}(\rho_0u_0^2+h_0^2)\right)v
\tag{3}
$$

$$
\frac{1}{\Delta t}\int h v + \nu \int h_{x} v_x = \frac{1}{\Delta t} \int h_0 v
\tag{4}
$$


Now take a finite subspace of the functions space with basis functions $\{\phi_i\}_{i=1}^n$ and define 
$$
\rho = \sum a_i \phi_i, \quad u = \sum b_i \phi_i, \quad h = \sum c_i \phi_i, \quad e = \sum d_i \phi_i
$$
and 
$$
\rho_0 = \sum a_i^0 \phi_i, \quad u_0 = \sum b_i^0 \phi_i, \quad h_0 = \sum c_i^0 \phi_i, \quad e_0 = \sum d_i^0 \phi_i
$$

Where each function without the subscript corresponds to the current time and those with correspond to the funcitons solved for in the previous time step.

Note also that $T$ is known.

$$
\frac{1}{\Delta t} \int \rho v - \int \rho u v_x = \frac{1}{\Delta t} \int \rho_0 v
$$

$$
\frac{1}{\Delta t} \int \rho v - \int \rho u^2 v_x +\mu \int u_x v_x = \frac{1}{\Delta t} \int \rho_0 u_0 v
$$

```python
# initialize mesh size
num_els = 1
x_vals,h = np.linspace(0,1,num_els+1,retstep=True)

# store coefficients
a_vals = []
a0_vals = []
b_vals = []
b0_vals = []
c_vals = []
c0_vals = []
d_vals = []
d0_vals = []

# create symbolic coefficients to solve for
for i in range(num_els+1):
    a_vals.append(sympify('a{}'.format(i)))
    a0_vals.append(sympify('a0{}'.format(i)))
    b_vals.append(sympify('b{}'.format(i)))
    b0_vals.append(sympify('b0{}'.format(i)))
    c_vals.append(sympify('c{}'.format(i)))
    c0_vals.append(sympify('c0{}'.format(i)))
    d_vals.append(sympify('d{}'.format(i)))
    d0_vals.append(sympify('d0{}'.format(i)))
```

```python
# create shape functions
shapes = []
doms = []

s1 = Piecewise((-x/h + 1, (0 <= x) & (x <= h)),(0, True))
shapes.append(s1)

for i in x_vals[1:-1]:
    s = Piecewise(((i-x)/h+1,(x<=i+h)&(x>=i)),((x-i)/h+1,(x<=i)&(x>=i-h)),(0,True))
    shapes.append(s)
    
sn = Piecewise(((x-1)/h+1,(x<=1)&(x>=1-h)),(0,True))
shapes.append(sn)

# create shape function derivatives
shape_ders = [diff(s,x) for s in shapes]
```

```python
# display the shape functions
domain = np.linspace(0,1,11)

for s in shapes:
    y = [s.subs(x,d) for d in domain]
    plt.plot(domain, y)
plt.title('Shape Functions')
plt.show()
```

```python
p, p0 = sympify(0), sympify(0)

u, u0 = sympify(0), sympify(0)

h, h0 = sympify(0), sympify(0)

e, e0 = sympify(0), sympify(0)


for i in range(num_els+1):
    p += a_vals[i]*shapes[i]
    p0 += a0_vals[i]*shapes[i]
    
    u += b_vals[i]*shapes[i]
    u0 += b0_vals[i]*shapes[i]
    
    h += c_vals[i]*shapes[i]
    h0 += c0_vals[i]*shapes[i]
    
    e += d_vals[i]*shapes[i]
    e0 += d0_vals[i]*shapes[i]
```

```python
# define remaining quantities
nu = 1/5
mu = 1
kap = 2
T = sympify(1)
dt = .01
```

```python
eqns1, eqns2, eqns3, eqns4 = [], [], [], []
    
for i in range(num_els+1):
    v = shapes[i]
    vp = shape_ders[i]
    eqns1.append(integrate(p*v,x)/dt - integrate(p*u*vp,x) - integrate(p0*v,x)/dt)
    eqns2.append(integrate(p*u*v,x)/dt - integrate((p*u**2+h**2)*vp,x) - 
                 integrate((p+h**2/2)*vp,x) + mu*integrate(diff(u,x)*vp,x)-integrate(p0*u0*v,x)/dt)
    eqns3.append(integrate((p*e+(p*u**2+h**2)/2)*v,x)/dt - integrate(((p*e+p*u**2/2+p)*u)*vp,x) +
                 integrate((nu+2*mu)*u*diff(u,x)*vp,x) + kap*integrate(diff(T,x)*vp,x) - 
                 integrate((p0*e0+(p0*u0**2+h0**2)/2)*v,x)/dt)
    eqns4.append(integrate(h*v,x)/dt + nu*integrate(diff(h,x)*vp,x) - integrate(h0*v,x)/dt)
    
eqns = [eqns1, eqns2, eqns3, eqns4]
```
